```python
import pandas as pd
import numpy as np

daily = pd.read_csv("FRED/Daily.csv")
daily_7 = pd.read_csv("FRED/Daily_7_Day.csv")
daily_close = pd.read_csv("FRED/Daily_Close.csv")
```

```python
daily.columns = ['Date', 'DGS1', 'DGS10', 'T10Y2Y', 'DBAA', 'USD1MTD156N', 'DCOILBRENTEU', 'DTWEXB', 'DEXUSEU', 'T10YIE', 'T5YIFR']
daily_7.columns = ['Date', 'DFF']
daily_close.columns = ['Date', 'BAMLH0A0HYM2EY']
```

```python
temp = pd.merge(daily, daily_7, on='Date', how='outer')
result = pd.merge(temp, daily_close, on='Date', how='outer')

print(result)
```
